## Definicja problemu
Celem ćwiczenia było stworzenie konfiguracji algorytmu ewolucyjnego dla biblioteki PyAge. Rozwiązywanym problemem był SAT-kCNF.

Dane wejściowe zaczerpnąłem z gotowych problemów CNF przygotowanych w formacie *DIMACS CNF* ze strony http://people.sc.fsu.edu/~jburkardt/data/cnf/cnf.html. Wybrałem problem z 1040 zmiennymi i 3668 klauzulami aby możliwe było zobrazowanie na wykresie postępu znajdywania wyniku przez algorytm.

### Genotyp
Genotyp dla problemu SAT CNF jest bardzo prosty - to zwykła lista wartości logicznych dla kolejnych zmiennych: [val_a, val_b, val_c, val_d, val_e, ...]. Podobnie jest również przedstawiona formuła logiczna problemu - jest to lista koniunkcyjnxg klauzul, które zawierają tuple alternatyw.

### Mutacje
* zamiana losowej ilości wartości logicznych genotypu na przeciwne
* przesunięcie bitowe wartości logicznych w lewo

### Crossowanie
Crossowanie przebiega tak, że z równym prawdopodobieństwem wykonujemy na obu bitach operacje OR bądź AND
